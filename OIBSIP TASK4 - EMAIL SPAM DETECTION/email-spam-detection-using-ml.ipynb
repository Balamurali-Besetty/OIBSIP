```python
import pandas as pd

file_path = '/kaggle/input/spam-mails-dataset/spam_ham_dataset.csv'
df = pd.read_csv(file_path, encoding='ISO-8859-1')
df.head()
```

```python
df['text'].value_counts()
```

```python
df['label'].value_counts()
```

```python
#separate x and y
x=df.text.values
y=df.label.values
```

```python
# split train and test
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size=.30)
```

```python
# Data preprocessing
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
X_train= cv.fit_transform(xtrain)
```

```python
X_train.toarray()
```

```python
# ML ALgorithm
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train,ytrain)
```

```python
x_test=cv.transform(xtest)
x_test.toarray()
```

```python
model.score(x_test, ytest)
```

****Model works out 97 percent****

```python
email = ['FREE IPHONE','IPHONE COSTS 60 THOUSAND','beleive me its way cheaper','rs 12000']
```

```python
cv_email = cv.transform(email)
model.predict(cv_email)
```
